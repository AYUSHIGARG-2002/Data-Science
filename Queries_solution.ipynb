```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
data = pd.read_csv("E:/AYUSHI/MICRO SQL/Taxi Riding Analysis/superstore.csv")

# Count the number of orders by Ship Mode
ship_mode_counts = data['Ship Mode'].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 6))
plt.pie(ship_mode_counts, labels=ship_mode_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Ship Mode Distribution')
plt.axis('equal')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
data = pd.read_csv("E:/AYUSHI/MICRO SQL/Taxi Riding Analysis/superstore.csv")

# Clean the Discount column
data['Discount'] = data['Discount'].str.replace('%', '').astype(float)

# Calculate the average discount by Region
average_discount_by_region = data.groupby('Region')['Discount'].mean()

# Plot the bar chart
plt.figure(figsize=(8, 6))
average_discount_by_region.plot(kind='bar')
plt.xlabel('Region')
plt.ylabel('Average Discount')
plt.title('Average Discount by Region')
plt.xticks(rotation=0)
plt.show()

# Find the Region with the highest average discount
region_highest_avg_discount = average_discount_by_region.idxmax()
print("Region with the highest average discount:", region_highest_avg_discount)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
data = pd.read_csv("E:/AYUSHI/MICRO SQL/Taxi Riding Analysis/superstore.csv")

# Group the data by Region and Segment and calculate the total sales
sales_by_region_segment = data.groupby(['Region', 'Segment'])['Sales'].sum().unstack()

# Plot the grouped bar chart
sales_by_region_segment.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.title('Sales by Region and Segment')
plt.xticks(rotation=0)
plt.legend(title='Segment')
plt.show()

```
